```python
bam_file = "./numts_test.bam"
chromo = "chrM"
with pysam.AlignmentFile(bam_file, "rb") as bam:
    for read in bam.fetch():
        primary_cigar = read.cigarstring
        primary_strand = "+" if read.is_reverse else "-"
        if read.has_tag('SA'):
            sa_tag = read.get_tag("SA")
            supplementary_positions = sa_tag.split(";")[:-1]
            for sa in supplementary_positions:
                chrom, pos, strand,cigar,mapq,nm = sa.split(',')
                if chrom != chromo:
                    print(chrom, pos, strand, cigar, primary_strand, primary_cigar, mapq, nm)
```
